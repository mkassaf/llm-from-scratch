# Chapter 3 — Attention Mechanisms

**Goal:** Understand the core mechanism that gives Transformers their power.

### Key concepts
- Dot-product attention: `Attention(Q,K,V) = softmax(QK^T / sqrt(d_k)) V`
- Causal mask (each token only sees past tokens)
- Multi-head attention (parallel attention heads)
- Attention weight visualization

```python
import sys, torch, math
import matplotlib.pyplot as plt
import torch.nn.functional as F
sys.path.insert(0, '..')
from src.model.attention import scaled_dot_product_attention, CausalSelfAttention, MultiHeadAttention
```

## 1. Scaled Dot-Product Attention from Scratch

```python
torch.manual_seed(42)
B, T, C = 1, 6, 8  # batch, seq_len, embed_dim

Q = torch.randn(B, 1, T, C)
K = torch.randn(B, 1, T, C)
V = torch.randn(B, 1, T, C)

out = scaled_dot_product_attention(Q, K, V)
print(f'Output shape: {out.shape}')   # (1, 1, 6, 8)
```

## 2. Visualise the Causal Mask

```python
T = 8
mask = torch.triu(torch.ones(T, T), diagonal=1).bool()

plt.figure(figsize=(5, 4))
plt.imshow(~mask.numpy(), cmap='Blues', vmin=0, vmax=1)
plt.title('Causal Attention Mask (blue = allowed)')
plt.xlabel('Key position'); plt.ylabel('Query position')
plt.colorbar(); plt.show()
```

## 3. Multi-Head Attention

```python
mha = MultiHeadAttention(embed_dim=64, num_heads=4, context_length=16)
x   = torch.randn(2, 10, 64)   # (batch=2, seq=10, embed=64)
out = mha(x)
print(f'MHA output shape: {out.shape}')   # (2, 10, 64)

params = sum(p.numel() for p in mha.parameters())
print(f'MHA parameters: {params:,}')
```
