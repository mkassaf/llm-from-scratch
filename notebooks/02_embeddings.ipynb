```python
import sys, torch
sys.path.insert(0, '..')
from src.data.dataset import GPTDataset, EmbeddingLayer, create_dataloader
from torch.utils.data import DataLoader
```

## 1. Sliding-Window Dataset

```python
# Fake token IDs to illustrate the concept
token_ids = list(range(20))
ds = GPTDataset(token_ids, context_length=4, stride=1)

print(f'Dataset size: {len(ds)} samples')
for i in range(3):
    x, y = ds[i]
    print(f'  Input:  {x.tolist()}')
    print(f'  Target: {y.tolist()}')  # shifted by 1
    print()
```

## 2. Embedding Layer

```python
VOCAB_SIZE     = 50257
EMBED_DIM      = 256
CONTEXT_LENGTH = 64

emb = EmbeddingLayer(VOCAB_SIZE, EMBED_DIM, CONTEXT_LENGTH)

batch = torch.randint(0, VOCAB_SIZE, (4, 16))  # (batch=4, seq=16)
out   = emb(batch)
print(f'Input shape:  {batch.shape}')
print(f'Output shape: {out.shape}')  # (4, 16, 256)
```

```python
# Visualize embedding magnitudes
import matplotlib.pyplot as plt

norms = out[0].norm(dim=-1).detach().numpy()  # one sample, seq norms
plt.figure(figsize=(8, 3))
plt.bar(range(len(norms)), norms)
plt.xlabel('Position'); plt.ylabel('Embedding L2 norm')
plt.title('Token embedding norms across sequence positions')
plt.show()
```
