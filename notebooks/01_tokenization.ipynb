# Chapter 2 — Tokenization

**Goal:** Convert raw text into integer token IDs that the model can process.

### Key concepts
- Vocabulary building
- Simple whitespace/punctuation tokenizer
- Byte-Pair Encoding (BPE) via `tiktoken`
- Special tokens: `<|unk|>`, `<|endoftext|>`

```python
import sys
sys.path.insert(0, '..')
from src.data.tokenizer import SimpleTokenizer, BPETokenizer
```

## 1. Simple Tokenizer

```python
text = """It was the best of times, it was the worst of times,
it was the age of wisdom, it was the age of foolishness."""

tokenizer = SimpleTokenizer.from_corpus(text)
print(f'Vocabulary size: {len(tokenizer.str_to_int)}')
print(f'Sample vocab: {list(tokenizer.str_to_int.items())[:10]}')
```

```python
sample = 'It was the best of times'
ids = tokenizer.encode(sample)
print('Encoded:', ids)
print('Decoded:', tokenizer.decode(ids))
```

## 2. BPE Tokenizer (tiktoken / GPT-2)

```python
bpe = BPETokenizer(encoding_name='gpt2')
print(f'BPE vocab size: {bpe.vocab_size:,}')

ids = bpe.encode(text)
print(f'Encoded ({len(ids)} tokens):', ids[:20], '...')
print('Decoded:', bpe.decode(ids[:20]))
```
