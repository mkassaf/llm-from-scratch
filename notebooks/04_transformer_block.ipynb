# Chapter 4 — Transformer Block

**Goal:** Assemble attention + feed-forward + residual connections into one Transformer block.

### Key concepts
- Pre-LayerNorm (GPT-style)
- Residual (skip) connections
- GELU activation
- Position-wise FFN (4× hidden dim)
- Dropout for regularisation

```python
import sys, torch
sys.path.insert(0, '..')
from src.model.transformer import TransformerBlock, GELU, FeedForward
```

## 1. GELU Activation

```python
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.functional as F

gelu = GELU()
x_vals = torch.linspace(-3, 3, 200)

plt.figure(figsize=(7, 3))
plt.plot(x_vals.numpy(), gelu(x_vals).numpy(), label='GELU')
plt.plot(x_vals.numpy(), F.relu(x_vals).numpy(), label='ReLU', linestyle='--')
plt.legend(); plt.title('GELU vs ReLU'); plt.grid(True); plt.show()
```

## 2. Feed-Forward Network

```python
ff = FeedForward(embed_dim=64, expansion=4)
x  = torch.randn(2, 10, 64)
print(f'FFN output shape: {ff(x).shape}')   # (2, 10, 64)
print(f'FFN parameters: {sum(p.numel() for p in ff.parameters()):,}')
```

## 3. Full Transformer Block

```python
block = TransformerBlock(embed_dim=64, num_heads=4, context_length=16, dropout=0.1)
x = torch.randn(2, 16, 64)
out = block(x)

print(f'Input  shape: {x.shape}')
print(f'Output shape: {out.shape}')   # same as input — residual connection preserves shape
print(f'Block parameters: {sum(p.numel() for p in block.parameters()):,}')
```
